```python
import pandas as pd
import pymysql
import math
import sys
import numpy as np 

import matplotlib.pyplot as plt
from future.utils import iteritems
from collections import Counter
import os
import csv
from konlpy.tag import Okt
twitter = Okt()

import matplotlib.font_manager as fm #한글이 나오게

#로컬 csv 읽기 csv형태 : 제목(0)/기사내용(1)/신문사(2)/카테고리(3)
os.chdir("###")
file = open('Article_Test.csv', 'r', encoding='euc-kr')#euc-kr
line = csv.reader(file)


##tf 1+log(tf)
def get_tf(csv, lank=6):
    spliter = Okt()
    nouns = spliter.nouns(csv)#'nouns'
    count = Counter(nouns)
    
    return_dict = {}  
    for n, c in count.most_common(lank):
        return_dict[n]=c  
    return return_dict

##idf

def get_idf(data1,data2,len):
    return_dict = {}
    
    score=get_score(data1,data2)#score dict 형태

    #idf=1
    for n in data1:
        #idf=math.log10((len-score.get(n))/score.get(n)) 
        
        idf=math.log10(len/score.get(n)) 
        return_dict[n]=idf
 
    return return_dict


#단어가 포함된 문서의수
def get_score(data1,data2):
    score_list={}
   
    for i in data1: 
        a=0    
        for n in range(len(data2)):
                    if i in data2[n]:
                       a+=1
        score_list[i]=a#자기 자신 제외하기
    return score_list



tf=[]
idf=[]

#tf_idf={}
tf_idf_list=[]

#카운트제외 명사만 저장
nouns_data=[]

#카운트 저장
nouns_dict=[]
a=0

# 기사내용으로 tf-idf
for i in line:
    tag=get_tf(i[1]) 
    tf.append(tag)
    
    #모든 기사의 명사
    spliter = Okt()
    nouns = spliter.nouns(i[1])
    count = Counter(nouns)
    nouns_data.append(list(count.keys()))
    nouns_dict.append(count)
   
     # if not tag:
    #          continue             
      #  except:
      #      pass

def main(argv):
    #print("모든 기사의 명사:",nouns_data)
    #print("0의 기사의 명사 갯수 확인:",nouns_dict[0])
    #print("value값 확인:",nouns_dict[0].get('비'))    
    #print("단어가포함된문서의수:",get_score(list(tf[0].keys()),nouns_data))
    #print("tf:",tf[0])
    #print("IDF값:",get_idf(list(tf[0].keys()),nouns_data,len(tf)))
    #idf.append(get_idf(list(tf[0].keys()),nouns_data,len(tf)))
    
    for i in nouns_data:
        idf.append(get_idf(i,nouns_data,len(nouns_data)))
        
   
    #print("IDF",idf[1])#숫자는 기사의 번호
    for data in nouns_data:
    
        tf_idf={}
        for i in range(len(data)):
            tf_idf[data[i]]=idf[nouns_data.index(data)].get(data[i])*nouns_dict[nouns_data.index(data)].get(data[i])
        tf_idf_list.append(tf_idf)
        
  
            #if not (idf[0].get(data[i])!=0):
            #    pass          

if __name__ == '__main__':
    main(sys.argv)

data = {'tf': nouns_dict[1], 
        'idf':idf[1], 
        'tf-idf':tf_idf_list[1],
        } 
df = pd.DataFrame(data)
df
df.sort_values(by=['tf-idf'], axis=0, ascending=False).head(6)
```
