```python
import os
import shutil
import glob
path_dir = '###' # 정리할 폴더
file_list = os.listdir(path_dir)        # path 에 존재하는 파일 목록 가져오기
file_list.sort()                        # 파일 이름 순서대로 정렬
#파일 jpg 삭제
#myExt = '*.jpg' # 확장자            
#files = glob.glob(path_dir+"/"+ myExt)
#for f in files:
#    os.remove(f)

remove_list=[]
for i in file_list:
    path_file=os.path.join(path_dir, i)
    file_count=len(os.listdir(path_file)) #각 폴더의 파일 개수 세기
    print(file_count)
    if file_count <= 20:
        remove_list.append(path_file)
        shutil.rmtree(path_file, ignore_errors=False, onerror=None) #20개이하 파일을 갖고있는 폴더 제거
print(remove_list)
print("완료")    
```
